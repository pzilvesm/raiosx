```python
import numpy as np
import math
import matplotlib.pyplot as plt
import copy
import pandas as pd
import os
from scipy.optimize import curve_fit
from scipy import optimize
from scipy.stats import poisson, norm
from scipy.signal import find_peaks
from matplotlib import colors
```

```python
data= open('AL2O3.txt', 'r') 
x = []
y = []
for line in data:
    p = line.split()
    angle = (p[0])
    psd = p[1]
    x.append(float(angle[:-1]))
    y.append(float(psd[:-1]))
```

```python
plt.rcParams.update({
"image.origin": "lower",
"image.aspect": 1,
"text.usetex": True,
"grid.alpha": .5,
}) 
plt.style.use('ggplot')
```

```python
peaks, _ = find_peaks(y, height=1000,threshold=90)
plt.figure(figsize=(10,5))
plt.plot(x,y,c='tab:gray')
plt.ylabel('PSD')
plt.xlabel('Angle')
plt.title('Al2O3')
for i in peaks: plt.plot(x[i],y[i],"x", label="peak angle: {:.2f}".format(x[i]))
plt.legend(fancybox=True, shadow=True,loc='upper left');
print(x[0],x[-1])
```

```python
angles=[math.radians(x[i]/2) for i in peaks]
dAA=[1.54184/(np.sin(j)*2) for j in angles]
print(dAA)
```

```python
data= open('AL2O3DET.txt', 'r') 
x = []
y = []
for line in data:
    p = line.split()
    angle = (p[0])
    psd = p[1]
    x.append(float(angle[:-1]))
    y.append(float(psd[:-1]))
```

```python
angles=[math.radians(x[i]/2) for i in peaks]
dAA=[1.54184/(np.sin(j)*2) for j in angles]
print(dAA)
```
